## 1. Bir Görüntü Aslında Bir Sayı Dizisidir

```python
import numpy as np
from skimage import data, color
import matplotlib.pyplot as plt

image = data.astronaut()
gray = color.rgb2gray(image)

print(f"Renkli görüntü şekli: {image.shape}")
print(f"Gri tonlama şekli: {gray.shape}")

plt.figure(figsize=(8,4))
plt.subplot(1,2,1); plt.imshow(image); plt.title("RGB"); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(gray, cmap='gray'); plt.title("Gri Tonlama"); plt.axis('off')
plt.show()
```

## 2. Sıfırdan Evrişim (Convolution)

```python
def convolve2d(image, kernel):
    img_h, img_w = image.shape
    k_h, k_w = kernel.shape
    out_h, out_w = img_h - k_h + 1, img_w - k_w + 1
    output = np.zeros((out_h, out_w))
    for i in range(out_h):
        for j in range(out_w):
            output[i, j] = np.sum(image[i:i+k_h, j:j+k_w] * kernel)
    return output

edge_kernel = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
small_gray = gray[::4, ::4]  # hızlı çalışması için küçült
edges = convolve2d(small_gray, edge_kernel)

plt.figure(figsize=(5,5))
plt.imshow(edges, cmap='gray')
plt.title("Kenar Tespiti")
plt.axis('off')
plt.show()
```

## 3. Deneyin!

Farklı çekirdekler (blur, sharpen) deneyin ve sonuçları karşılaştırın.

```python
# Kendi çekirdeğinizi tasarlayın
my_kernel = np.ones((3,3)) / 9  # blur
result = convolve2d(small_gray, my_kernel)
plt.imshow(result, cmap='gray'); plt.axis('off'); plt.show()
```
