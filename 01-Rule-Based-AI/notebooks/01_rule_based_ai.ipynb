## 1. Hava Durumu Asistanı

```python
from dataclasses import dataclass

@dataclass
class WeatherReading:
    temperature_c: float
    is_raining: bool
    wind_speed_kmh: float

def get_weather_advice(reading: WeatherReading) -> list[str]:
    advice = []
    if reading.temperature_c <= 0:
        advice.append("🥶 Hava dondurucu. Kalın mont giyin.")
    elif reading.temperature_c <= 15:
        advice.append("🧥 Hava soğuk. Ceket önerilir.")
    elif reading.temperature_c <= 25:
        advice.append("🙂 Hava ılıman.")
    else:
        advice.append("🥵 Hava sıcak.")
    if reading.is_raining:
        advice.append("☔ Şemsiye alın.")
    return advice

get_weather_advice(WeatherReading(temperature_c=5, is_raining=True, wind_speed_kmh=10))
```

## 2. Deneyin!

Yukarıdaki `WeatherReading` değerlerini değiştirip tekrar çalıştırın. Farklı sıcaklık ve yağış kombinasyonlarında hangi kuralların tetiklendiğini gözlemleyin.

```python
# Kendi denemenizi buraya yazın
get_weather_advice(WeatherReading(temperature_c=32, is_raining=False, wind_speed_kmh=5))
```
